# Gradient Descent
 is an optimization algorithm used to minimize a loss (error) function by updating model parameters step-by-step in the direction that reduces error.

# Simple analogy
**Imagine you are standing on a mountain in the dark and want to reach the lowest valley:**



1.   Move in the steepest downward direction
2.   Repeat until you reach the bottom

# That “steepest downward direction” is the gradient.

# Mathematical Definition

θ = θ − α∇J(θ)

where


1. θ = model parameters (weights)
2. α = learning rate
3. ∇J(θ) = gradient of loss function


# Why Gradient Descent is Needed
# In ML we train models by minimizing a loss function:



1.   Linear Regression → Mean Squared Error

2.   Logistic Regression → Log Loss
3.   Neural Networks → Cross Entropy

Without gradient descent:
1.  We cannot train deep networks
2. Closed-form solutions (like normal equation) become impossible



# How Gradient Descent Works Step-by-Step
1. Initialize weights randomly
2. Compute predictions
3. Calculate loss
4. Compute gradient (derivative)
5. Update weights
6. Repeat until convergence

```python
import numpy as np

X = np.array([1,2,3])
y = np.array([2,4,6])

# Step 1: Initialize weights randomly
w = np.random.randn()

lr = 0.01

for epoch in range(100):

    # Step 2: Compute predictions
    y_pred = w * X

    # Step 3: Calculate loss
    loss = np.mean((y - y_pred)**2)

    # Step 4: Compute gradient
    grad = -2*np.mean(X*(y - y_pred))

    # Step 5: Update weights
    w -= lr * grad

    # Step 6: Repeat until convergence
```

# Types of Gradient Descent

# 1. Batch Gradient Descent
*  θ = θ - α * gradient(full dataset)
*  Stable
*  Accurate gradient
*  very slow on large datasets


# 2. Stochastic Gradient Descent (SGD)
*  for each sample:

    θ = θ - α * gradient(sample)

* fast
* Escapest local minima

* Noisy updates


# 3. Mini-Batch Gradient Descent (Most used)

* Uses small batches (e.g., 32, 64, 128 samples)

* This is what deep learning frameworks use.

* why best  it combines
   * speed of SGD
   * stability of Batch GD

```python
for epoch in range(100):

    # Step 2: Compute predictions
    y_pred = w * X

    # Step 3: Calculate loss
    loss = np.mean((y - y_pred)**2)

    # Step 4: Compute gradient using full dataset
    grad = -2*np.mean(X*(y - y_pred))

    # Step 5: Update weights
    w -= lr * grad
```

```python
for epoch in range(100):
    for i in range(len(X)):

        # Step 2: Compute predictions
        y_pred = w * X[i]

        # Step 3: Calculate loss
        loss = (y[i] - y_pred)**2

        # Step 4: Compute gradient
        grad = -2*X[i]*(y[i] - y_pred)

        # Step 5: Update weights
        w -= lr * grad
```

```python
batch_size = 2

for epoch in range(100):
    for i in range(0, len(X), batch_size):

        xb = X[i:i+batch_size]
        yb = y[i:i+batch_size]

        # Step 2: Compute predictions
        y_pred = w * xb

        # Step 3: Calculate loss
        loss = np.mean((yb - y_pred)**2)

        # Step 4: Compute gradient
        grad = -2*np.mean(xb*(yb - y_pred))

        # Step 5: Update weights
        w -= lr * grad
```

# Deep Learning — PyTorch Backprop + GD



```python
import torch
import torch.nn as nn

model = nn.Linear(3,1)

optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
loss_fn = nn.MSELoss()

for epoch in range(50):

    # Step 2: Compute predictions
    pred = model(X)

    # Step 3: Calculate loss
    loss = loss_fn(pred, y)

    # Step 4: Compute gradient
    loss.backward()

    # Step 5: Update weights
    optimizer.step()

    # Step 6: Repeat
    optimizer.zero_grad()
```
