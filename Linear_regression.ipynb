```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# linear regression used to predict the continuous calues

# equation y = mx +b here m is slop b is intercept

# goal of LR is find the best fit line

```python
data = {
    "experience": [1, 2, 3, 4, 5],
    "salary": [3, 4, 6, 8, 10]
}
```

```python
from sklearn.linear_model import LinearRegression
```

```python
df = pd.DataFrame(data)
```

```python
X = df[['experience']]
y = df['salary']
```

```python
model = LinearRegression()
```

```python
plt.scatter(X,y)
```

```python
model.fit(X,y)
```

```python
pred =model.predict(X)
```

```python

print("Predictions:", pred)
```

```python

print("Salary for 6 years:", model.predict([[6]]))
```

```python
plt.scatter(X,y)
plt.plot(X,model.predict(X))
plt.show()
```

# Metrics for LR


1.  Mean Absolute Error (MAE)
This measures average absolute difference between actual and predicted values.
If MAE is small → model is good.

2.   Mean Squared Error (MSE)
Used internally by linear regression to learn the best line.
3.   Root Mean Squared Error (RMSE)


4.   R² Score (Most Important)
This tells how much variance in data is explained by the model



```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

# | Metric | What it tells            |
| ------ | ------------------------ |
| MAE    | Average error            |
| MSE    | Penalizes large mistakes |
| RMSE   | Error in real units      |
| R²     | Overall model quality    |


```python
import math
mae = mean_absolute_error(y, pred)
mse = mean_squared_error(y,pred)
rmse = math.sqrt(mse)
r2 = r2_score(y,pred)
```

```python
print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("R2 Score:", r2)
```
