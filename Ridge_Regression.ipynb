# . Why Ridge is Used



| Problem                    | Ridge Solution                 |
| -------------------------- | ------------------------------ |
| Overfitting                | Shrinks coefficients           |
| Highly correlated features | Distributes weight across them |
| Large variance model       | Adds bias → reduces variance   |




# Important Hyperparameters and Their Meaning

1. alpha (MOST IMPORTANT)

| Alpha value | Effect                    |
| ----------- | ------------------------- |
| 0           | Same as Linear Regression |
| Small       | Slight regularization     |
| Large       | Strong shrinkage          |


```python
alpha = 10
```

# solver

* Solver decides how the optimization is solved

| Solver   | Use case                        |
| -------- | ------------------------------- |
| auto     | sklearn chooses                 |
| svd      | very stable but slow            |
| cholesky | fast for small data             |
| sag      | large dataset                   |
| **saga** | large + sparse + regularization |


```python
solver = 'saga'
```

# tol (tolerance)

* Stopping criteria of solver.

| tol   | effect                |
| ----- | --------------------- |
| small | more precise, slower  |
| large | faster, less accurate |




```python
tol = 1e-4
```
